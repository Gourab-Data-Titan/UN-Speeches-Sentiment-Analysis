```python
# !pip install gensim spacy

import pandas as pd
import gensim
from gensim.models.ldamodel import LdaModel
from gensim.models import CoherenceModel
from gensim.corpora.dictionary import Dictionary
import matplotlib.pyplot as plt
df = filtered_df


# Step 1: Preprocessing - Tokenization
texts =df['filtered_text'].apply(lambda x: word_tokenize(x.lower()))

# Step 2: Creating the Dictionary and Corpus
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Step 3: Determine Coherence Scores
def compute_coherence_values(corpus, dictionary, texts, k_values, coherence='c_v'):
    coherence_values = []
    
    for k in k_values:
        lda_model = LdaModel(corpus=corpus, num_topics=k, id2word=dictionary, random_state=42)
        coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence=coherence)
        coherence_values.append(coherence_model_lda.get_coherence())
        
    return coherence_values

# Step 4: Plotting Coherence Scores
def plot_coherence_graph(k_values, coherence_values, coherence_type):
    plt.plot(k_values, coherence_values)
    plt.xlabel("Number of Topics")
    plt.ylabel("Coherence Score")
    plt.title(f"Coherence Score using {coherence_type} Coherence")
    plt.show()

# Step 5: Test with a range of topics and plot the coherence score graph for each coherence type
k_values = range(2, 11)  # Choose the range of topics you want to test (2 to 10 here)

# Coherence Types
coherence_types = ['u_mass', 'c_v', 'c_uci', 'c_npmi']

for coherence in coherence_types:
    coherence_scores = compute_coherence_values(corpus, dictionary, texts, k_values, coherence=coherence)
    print(f"Coherence values for {coherence}: {coherence_scores}")
    plot_coherence_graph(k_values, coherence_scores, coherence)

```

```python
import pandas as pd
import gensim
from gensim import corpora
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


nltk.download('stopwords')
nltk.download('punkt')

df = filtered_df

# Preprocess the text documents
stop_words = set(stopwords.words('english'))

def preprocess(doc):
    tokens = word_tokenize(doc.lower())
    return [word for word in tokens if word.isalnum() and word not in stop_words]

df['processed_text'] = df['filtered_text'].apply(preprocess)

# Create a dictionary and corpus needed for the LDA model
texts = df['processed_text'].tolist()
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Function to compute coherence scores for multiple coherence metrics
def compute_coherence_values(dictionary, corpus, texts, start=2, limit=10, step=1):
    coherence_values_umass = []
    coherence_values_cv = []
    coherence_values_cuci = []
    coherence_values_cnpmi = []
    model_list = []
    
    for num_topics in range(start, limit, step):
        # Train LDA model
        model = gensim.models.ldamodel.LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=10)
        model_list.append(model)
        
        # Coherence values for different algorithms
        coherence_model_umass = CoherenceModel(model=model, corpus=corpus, dictionary=dictionary, coherence='u_mass')
        coherence_model_cv = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_model_cuci = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_uci')
        coherence_model_cnpmi = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_npmi')
        
        coherence_values_umass.append(coherence_model_umass.get_coherence())
        coherence_values_cv.append(coherence_model_cv.get_coherence())
        coherence_values_cuci.append(coherence_model_cuci.get_coherence())
        coherence_values_cnpmi.append(coherence_model_cnpmi.get_coherence())
        
    return model_list, coherence_values_umass, coherence_values_cv, coherence_values_cuci, coherence_values_cnpmi

# Set the range for the number of topics to test
start = 2
limit = 12  # Test for topics from 2 to 11
step = 1

# Compute coherence scores
model_list, coherence_values_umass, coherence_values_cv, coherence_values_cuci, coherence_values_cnpmi = compute_coherence_values(dictionary, corpus, texts, start, limit, step)

# Visualize the coherence scores
x = range(start, limit, step)

plt.figure(figsize=(10, 6))
plt.plot(x, coherence_values_umass, label="u_mass", color='red', marker='o')
plt.plot(x, coherence_values_cv, label="c_v", color='blue', marker='o')
plt.plot(x, coherence_values_cuci, label="c_uci", color='green', marker='o')
plt.plot(x, coherence_values_cnpmi, label="c_npmi", color='orange', marker='o')

plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.title("Coherence Scores for Different Numbers of Topics")
plt.legend()
plt.grid()
plt.show()

# Find and print the optimal number of topics based on the highest coherence score for each algorithm
best_num_topics_umass = x[coherence_values_umass.index(max(coherence_values_umass))]
best_num_topics_cv = x[coherence_values_cv.index(max(coherence_values_cv))]
best_num_topics_cuci = x[coherence_values_cuci.index(max(coherence_values_cuci))]
best_num_topics_cnpmi = x[coherence_values_cnpmi.index(max(coherence_values_cnpmi))]

print(f"Optimal number of topics based on u_mass: {best_num_topics_umass}")
print(f"Optimal number of topics based on c_v: {best_num_topics_cv}")
print(f"Optimal number of topics based on c_uci: {best_num_topics_cuci}")
print(f"Optimal number of topics based on c_npmi: {best_num_topics_cnpmi}")
```

```python
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation as LDA
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Create a DataFrame with sample text data
df = filtered_df

# Step 2: Convert the text data into a document-term matrix using CountVectorizer
count_vectorizer = CountVectorizer(max_features=5000, stop_words='english')
doc_term_matrix = count_vectorizer.fit_transform(df['filtered_text'])

# Step 3: Apply LDA (Latent Dirichlet Allocation) for topic modeling
lda_model = LDA(n_components=3, random_state=42)  # Using 3 topics for simplicity
lda_topics = lda_model.fit_transform(doc_term_matrix)

# Step 4: Label each document with its dominant topic
dominant_topic_labels = np.argmax(lda_topics, axis=1)

# Add the dominant topic label to the DataFrame
df['dominant_topic'] = dominant_topic_labels

# Step 5: Display the contribution of all topics in each document (Numerically)
def display_numerical_contributions(df, lda_topics, num_topics):
    print("\nContribution of Topics for Each Document:")
    for index, row in df.iterrows():
        print(f"\nDocument #{index + 1}: {df['filtered_text'][index]}")
        for topic_idx in range(num_topics):
            print(f"Topic {topic_idx + 1}: {lda_topics[index, topic_idx]:.4f}")

display_numerical_contributions(df, lda_topics, lda_model.n_components)

# Step 6: Visualize the contribution of all topics in each document (Visually) using a stacked bar chart
def plot_topic_contributions(lda_topics, num_topics):
    plt.figure(figsize=(10, 6))
    
    # Prepare the data for stacked bar chart
    topic_contrib_df = pd.DataFrame(lda_topics, columns=[f'Topic {i+1}' for i in range(num_topics)])
    topic_contrib_df.plot(kind='bar', stacked=True, colormap='tab20', figsize=(12, 8))
    
    plt.title('Contribution of Topics in Each Document')
    plt.xlabel('Document Index')
    plt.ylabel('Topic Contribution')
    plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
    plt.tight_layout()
    plt.show()

# Step 7: Call the function to visualize the contributions
plot_topic_contributions(lda_topics, lda_model.n_components)

# Step 8: Visualize the contribution as a heatmap for another representation
def plot_heatmap(lda_topics):
    plt.figure(figsize=(10, 6))
    sns.heatmap(lda_topics, cmap='Blues', annot=True, cbar=True, xticklabels=[f'Topic {i+1}' for i in range(lda_model.n_components)])
    plt.title('Heatmap of Topic Contributions in Each Document')
    plt.xlabel('Topic')
    plt.ylabel('Document Index')
    plt.tight_layout()
    plt.show()

# Step 9: Call the heatmap function to visualize topic contributions
plot_heatmap(lda_topics)
```

```python
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation as LDA
from sklearn.feature_extraction.text import CountVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Ensure necessary nltk packages are downloaded
nltk.download('vader_lexicon')

# Step 1
df = filtered_df

# Step 2: Perform Sentiment Analysis using VADER
vader_analyzer = SentimentIntensityAnalyzer()

def get_sentiment_vader(text):
    """Classifies sentiment as positive, negative, or neutral using VADER."""
    scores = vader_analyzer.polarity_scores(text)
    if scores['compound'] >= 0.05:
        return 'positive'
    elif scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment'] = df['filtered_text'].apply(get_sentiment_vader)

# Step 3: Perform Emotion Analysis using heuristic rules
# We will map sentiment to certain emotions (this can be expanded with more rules or a pre-trained model)
def get_emotion_by_sentiment(sentiment):
    """Map sentiment to a basic set of emotions based on heuristic rules."""
    emotion_mapping = {
        'positive': ['joy', 'trust', 'optimism', 'love'],
        'negative': ['anger', 'sadness', 'fear', 'disgust', 'pessimism'],
        'neutral': ['neutral', 'anticipation', 'surprise']
    }
    
    if sentiment == 'positive':
        return emotion_mapping['positive'][0]  # Assign joy as default positive emotion
    elif sentiment == 'negative':
        return emotion_mapping['negative'][0]  # Assign anger as default negative emotion
    else:
        return emotion_mapping['neutral'][0]  # Assign neutral as default emotion

df['emotion'] = df['sentiment'].apply(get_emotion_by_sentiment)

# Step 4: Convert the text data into a document-term matrix using CountVectorizer
count_vectorizer = CountVectorizer(max_features=5000, stop_words='english')
doc_term_matrix = count_vectorizer.fit_transform(df["filtered_text"])

# Step 5: Apply LDA (Latent Dirichlet Allocation) for topic modeling
lda_model = LDA(n_components=3, random_state=42)  # Adjust number of topics as needed
lda_topics = lda_model.fit_transform(doc_term_matrix)

# Step 6: Label each document with its dominant topic
dominant_topic_labels = lda_topics.argmax(axis=1)
df['dominant_topic'] = dominant_topic_labels

# Step 7: Display annotated DataFrame
print("Annotated DataFrame:")
print(df)

# Optional: Visualize the sentiment and emotion distribution across topics
import matplotlib.pyplot as plt
import seaborn as sns

# Count of sentiment across topics
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='dominant_topic', hue='sentiment')
plt.title('Sentiment Distribution Across Topics')
plt.show()

# Count of emotion across topics
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='dominant_topic', hue='emotion')
plt.title('Emotion Distribution Across Topics')
plt.show()
```

```python

import os
import pandas as pd
import string
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to remove punctuation and convert text to lowercase
def preprocess_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Convert text to lowercase
    text = text.lower()
    return text

```

```python
# Directory containing my documents
directory = 'C:/Users/goura/OneDrive/Desktop/Project_leadership/Text_Analytics/Session 56-71/Session 69 - 2014'

# List to store preprocessed documents
preprocessed_documents = []

# Read and preprocess documents
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            # Read the file content
            content = file.read()
            # Preprocess the content (remove punctuation and convert to lowercase)
            cleaned_content = preprocess_text(content)
            # Append the cleaned content
            preprocessed_documents.append(cleaned_content)

# Combine all documents into a single string
all_text = ' '.join(preprocessed_documents)

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Plot the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.show()
```
